```python
import numpy as np
```

```python
data=np.loadtxt("lab7_ex2_lp_ip_coef.txt",dtype='str')
```

```python
data
```

```python
model=ConcreteModel()
```

```python
data.shape
```

```python
len(data[0])
```

```python
obj_coef_c = data[0].split(',')
```

```python
len(data[1].split(','))
```

```python
constraint_coef=[]
for i in data[1:]:
  constraint_coef.append(i.split(','))
```

```python
len(constraint_coef)
```

```python
constr_rhs_b=[]
for i in constraint_coef:
  constr_rhs_b.append(i[500])
```

```python
constr_rhs_b
```

```python
N=5
M=501
row_indices=np.arange(N)
col_indices=np.arange(M)
```

```python
print(row_indices)
```

```python
model.x = Var(col_indices[:-1], domain=Binary)
```

```python
model.x
```

```python
model.constraints = ConstraintList()
```

```python
for i in row_indices:
    model.constraints.add(sum(int(constraint_coef[i][j])*model.x[j] for j in col_indices[:-1]) <= int(constr_rhs_b[i]))
```

```python
model.objective = Objective(expr= sum(int(obj_coef_c[i])*model.x[i] for i in col_indices[:-1]), sense=maximize)
```

```python
model.pprint()
```

```python
!apt-get install -y -qq coinor-cbc
```

```python
opt_cbc = SolverFactory('cbc')
```

```python
result = opt_cbc.solve(model)
print('Solver status:', result.solver.status)
print('Solver termination condition:',result.solver.termination_condition)
print('Solver time:',result.solver.time)
```

```python
print('Branch and BOund Statistics',result.solver.statistics)
```

```python
print(result)
```

```python
print('\nObjective = ', model.objective())
```

Q3:
We are to change domain now.first we took domain binary now we take domain [0,1] interval.

```python
model.x.domain=Reals
for i in col_indices[:-1]:
   model.x[i].setlb(0)
   model.x[i].setub(1)
```

```python
result = opt_cbc.solve(model)
print('Solver status:', result.solver.status)
print('Solver termination condition:',result.solver.termination_condition)
print('Solver time:',result.solver.time)
```

```python
print(result)
```

```python
print('\nObjective = ', model.objective())
```

Here solver time is less than the condition when domain was binary since it takes less time to check because when domain is binary it is restricted  to take only 0 and 1 value but in interval it can take any value in interval.
