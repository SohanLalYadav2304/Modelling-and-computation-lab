Answer1:
CityScape Constructions Inc. has to build n ∈ N different types of factories at n type of different locations.Assume that variable $ x_{ij}  $ denotes ith factory and jth loaction.
If $x_{ij}$ = 1 it means that $i^{th}$ factory is located at $j^{th}$ location.
and if $ x_{ij}$ =0 it means that  $i^{th}$ factory is not located at $j^{th}$ location.
So our variable takes values as binary.

Since it is given that each factory must be located at one location so:
$ \sum_{j=1}^nx_{ij}=1  $ for all i=1,2,$\cdots$n

We are also given that each location must be alloted a facory to be located so:
$ \sum_{i=1}^nx_{ij}=1  $ for all j=1,2,$\cdots ,n$ \\
$ x_{ij}\geq 0 $ ∀ i,j∈{1,2,3,...,n}

Let we have a matrix A$_{ij}$ = [$a_{ij}$] $ ∀ i,j \in \{1,2,3,\cdots,n\} $ and $a_{ij}$ denotes the cost to lacate $i^{th}$ factory at $j^{th}$ location.
Objective function: $ A_{ij} $[x$_{ij}$] 

We are to minimize this cost function.

```python
opt_cbc = SolverFactory('cbc')
```

First we define a function for n,data,d for assignment of factories to locations. Here n is the number of factories and corresponding number of locations.And data is representing the array obtained from .txt file which contains the cost for allocation of factories to various lacations and d represents the domain of variables. 

```python
def assignment(n,data,d):
  row_indices=range(n)
  col_indices=range(n)
  model=ConcreteModel()
  model.x=Var(row_indices,col_indices,domain=d)
  model.objective = Objective(expr=sum(int(data[i+1,j+1])*model.x[i,j] for i in row_indices for j in col_indices),sense=minimize)
  model.constraints = ConstraintList()
  for j in range(n):
    model.constraints.add(expr=sum(model.x[i,j] for i in row_indices) == 1)
  for i in range(n):
    model.constraints.add(expr=sum(model.x[i,j] for j in col_indices) == 1)  
  result = opt_cbc.solve(model)
  print('Solver status:', result.solver.status)
  print('Solver termination condition:',result.solver.termination_condition) 
  print('\nObjective = ', model.objective())

  print('\nDecision Variables')
  for i in row_indices:
    for j in col_indices:
      print(f'whether factory {i+1} is assigned to location {j+1}:', model.x[i,j].value)

  print('\nConstraints')
  model.constraints.display()
  for i in range(n):
    for j in range(n):
      if model.x[i,j].value == 1:
        print(f'factory {i+1} is assigned to location {j+1}')
        
    
      
    
```
