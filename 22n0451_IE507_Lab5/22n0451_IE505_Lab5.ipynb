```python
import pandas as pd
import numpy as np 
```

```python
df=pd.read_csv("Berlin-tram-route - Sheet1.csv")
df
```

```python
new_df=df.replace(np.NaN,0)
```

```python
n_column = list(new_df.columns[1:]) 
while len(n_column)>0:
  group = [n_column[0]]
  n_column.remove(group[0])
  k = n_column.copy()
  for i in k:
    k =0
    for j in group:
      if new_df[new_df["route"]==j][i].values[0]==0:
        k = k+1
    if k == len(group):
      group.append(i)
      n_column.remove(i)
  print(group)
```

Here we get 7 groups of routes so we are needed 7 colours to show routes on map.
